```python
import numpy as np
import pandas as pd
from datetime import datetime

## 讀入資料
from google.colab import drive
drive.mount("/content/gdrive")
df=pd.read_csv('/content/gdrive/MyDrive/專案code/Cigarette_Brand/188Category.csv',encoding="cp950")
df1 =df.sort_values(['xdid', 'invDate'], ascending = [True, True])

## 去掉沒必要的欄位
df1=df1.drop(labels=['category', 'invtime'],axis=1)  # axis=1 表示按column除


## 新增週間/週末欄位
df1['週間']=df1['週末']=0

### 週間/週末
for index in range(len(df1)):
  temp = pd.Timestamp(str(df1.invDate[index]))
  ## print(index, df1.invDate[index], temp.dayofweek+1)
  if ((temp.dayofweek+1)<6):
    df1.週間[index]=1
  else:
    df1.週末[index]=1

df1['filter_tag'] =0
df1.filter_tag[0]=1


for index in range(len(df1)-1):
  if (df1.invNum[index+1]== df1.invNum[index]):
    df1.filter_tag[index+1]=0
  else:
    df1.filter_tag[index+1]=1

df2=df1[df1['filter_tag'] ==1]

df2= df2.drop(labels=['filter_tag'] ,axis=1)

## 無小計輸出到 csv 檔案
df2.to_csv('/content/gdrive/MyDrive/專案code/Cigarette_Brand/NotCount_WeekdayWeekend.csv',encoding = "cp950", index=False)

## groupby xdid
df3= df2.groupby(['xdid']).sum()


## 小計輸出到 csv 檔案
df3.reset_index().to_csv ('/content/gdrive/MyDrive/專案code/Cigarette_Brand/Count_WeekdayWeekend.csv',encoding = "cp950", index=False)
```
