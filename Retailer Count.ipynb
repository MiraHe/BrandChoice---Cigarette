```python
import numpy as np
import pandas as pd
from datetime import datetime

## 讀入資料
from google.colab import drive
drive.mount("/content/gdrive")
df = pd.read_excel('/content/gdrive/MyDrive/專案code/Cigarette_Brand/ OriginalData.xlsx')

df= df[df['category'] == 'Cigarette']   ## 篩出 ['category'] = 'Cigarette'
df1= df.groupby(['retailer'])['retailer'].size().reset_index(name='counts')
df1 =df1.sort_values(['counts'], ascending = [False])

df1['retailer'].iloc[0:6]

## 知道後   回去df 撈取
df2= df[(df['retailer'] == df1['retailer'].iloc[0]) | (df['retailer'] == df1['retailer'].iloc[1]) | (df['retailer'] == df1['retailer'].iloc[2]) | (df['retailer'] == df1['retailer'].iloc[3]) | (df['retailer'] == df1['retailer'].iloc[4]) | (df['retailer'] == df1['retailer'].iloc[5])]

## df3=df2[['xdid', 'invNum', 'invDate', 'invtime', 'retailer']]
df3=df2[['xdid', 'invNum', 'retailer']]

##  將retailer 轉換成每行的 dummy variable
df4=pd.get_dummies(df3, columns = ['retailer'])

df4 =df4.sort_values(['invNum'], ascending = [True])
df4=df4.reset_index(drop=True)

df4['filter_tag'] =0
df4.filter_tag[0]=1

for index in range(len(df4)-1):
  if (df4.invNum[index+1]== df4.invNum[index]):
    df4.filter_tag[index+1]=0
  else:
    df4.filter_tag[index+1]=1

df5=df4[df4['filter_tag'] ==1]

## 去掉沒必要的欄位
df6= df5.drop(labels=['filter_tag'] ,axis=1)

## 無小計輸出到 csv 檔案
df6.to_csv('/content/gdrive/MyDrive/專案code/Cigarette_Brand/retailer_not_count.csv', encoding="UTF-8", index=False)

## 小計
df7 = df6.groupby(['xdid']).sum()

## 小計輸出到 csv 檔案
df7.reset_index().to_csv ('/content/gdrive/MyDrive/專案code/Cigarette_Brand/retailer_count.csv', encoding="UTF-8", index=False)
```
