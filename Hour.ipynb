```python
import numpy as np
import pandas as pd
from datetime import datetime

## 讀入資料
from google.colab import drive
drive.mount("/content/gdrive")
df = pd.read_csv('/content/gdrive/MyDrive/專案code/Cigarette_Brand/188Category.csv',encoding="cp950")
df1 =df.sort_values(['xdid', 'invDate'], ascending = [True, True])


## 去掉沒必要的欄位
df1=df1.drop(labels=['category', 'invDate',],axis=1)  # axis=1 表示按column?除

## 新增hour欄位
df1['早上']=df1['中午']=df1['晚上']=df1['半夜']=0

### hour
for index in range(len(df1)):
  temp = int(str(df1.invtime[index])[0:2])
  if   temp<6:
    df1.半夜[index]=1
  elif  temp<12:
    df1.早上[index]=1
  elif  temp<18:
    df1.中午[index]=1
  else:
    df1.晚上[index]=1

df1['filter_tag'] =0
df1.filter_tag[0]=1


for index in range(len(df1)-1):
  if (df1.invNum[index+1]== df1.invNum[index]):
    df1.filter_tag[index+1]=0
  else:
    df1.filter_tag[index+1]=1

df2=df1[df1['filter_tag'] ==1]

df2= df2.drop(labels=['filter_tag'] ,axis=1)


## 無小計輸出到 csv 檔案
df2.to_csv('/content/gdrive/MyDrive/專案code/Cigarette_Brand/hour_not_count.csv', encoding="cp950", index=False)

## groupby xdid
df3= df2.groupby(['xdid']).sum()

## 小計輸出到 csv 檔案
df3.reset_index().to_csv ('/content/gdrive/MyDrive/專案code/Cigarette_Brand/hourt_count.csv', encoding="cp950", index=False)
```
